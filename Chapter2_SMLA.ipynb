```python
train = pd.read_csv(r'train.csv')
train
```

Eliminamos todos los **NaN** para KNeighborsClassifier(n_neighbors=3)

```python
train = train.fillna(0)
train
```

Convertimos la columna **Transported** en False=0 y True=1

```python
train["Transported"] = train["Transported"].astype(int)
```

```python
train.iloc[:, [5,7,8]]
```

```python
train.iloc[:, [5]]
```

```python
X = train.iloc[:, [5,7,8]]
y = train['Transported']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,random_state = 0)
```

```python
from sklearn.neighbors import KNeighborsClassifier
clf = KNeighborsClassifier(n_neighbors=3)
```

```python
y_train
```

```python
X_train
```

```python
clf.fit(X_train,y_train)
```

```python
print("Test set predictions: {}".format(clf.predict(X_test)))
```

```python
print("Test set accuracy: {:.2f}".format(clf.score(X_test,y_test)))
```

```python
data = list(zip(X_test['Age'], X_test['RoomService']))
print(data)
```

```python
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(data, y_test)
```

Use the same KNN object to predict the class of new, unforeseen data points. First we create new **x** and **y** features and then call *knn.predict()** on the new data point to get a class of 0 "False" or 1 "True".

```python
new_x = 8
new_y = 21
new_point = [(new_x, new_y)]
prediction = knn.predict(new_point)
print(prediction)
```

```python
plt.scatter(X_test['Age'] + [new_x], X_test['RoomService'] + [new_y], c=y_test + [prediction[0]])
plt.text(x=new_x-1.7, y=new_y-0.7, s=f"new point, class: {prediction[0]}")
plt.show()
```
