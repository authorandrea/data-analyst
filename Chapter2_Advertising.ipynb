```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv(r'Advertising.csv')
print(df)
```

```python
#Scatter x(TV) y y(Sales)
plt.scatter(df['TV'], df['sales'])

plt.xlabel('TV')
plt.ylabel('sales')
```

```python
#plot Radio vs Sales DataFrames

df.plot.scatter(x="radio", y="sales", alpha=0.5, color = 'red')
plt.show()
```

```python
#plot newspaper vs Sales DataFrames

df.plot.scatter(x="newspaper", y="sales", alpha=0.5, color = 'red')
plt.show()
```
